Create Fake Data for performance test in SSIS / SQL Server

```python
import pandas as pd
from faker import Faker
import datetime

fake = Faker('de_DE')
column_names = ["customer_name", "country", "email", "website"]
```

```python


data = [(fake.name(), fake.country(), fake.email(), fake.url()) for i in range(10000)]

data = pd.DataFrame(data, columns=column_names)
data.insert(0,'customer_id', data.index)



data
```

```python
#Save full DataSet and Half dataset

data.to_csv('full_data.csv', index=False)
data[:5000].to_csv('half_data.csv', index=False)
```

```python
data_new = [(fake.name(), fake.country(), fake.email(), fake.url()) for i in range(5000)]
data_new = pd.DataFrame(data_new, columns=column_names)
data_new.insert(0, 'customer_id', data_new.index)

data_new_upper_half = pd.concat([data_new, data[5000:]])

data_new_upper_half.to_csv("half_data_updates.csv", index=False)
```

```python
#Umrechnung von Start und Endzeit des SSIS Prozesses in Sekunden

start = '10:48:12:084'
finish = '10:48:33:064'

start_split = start.split(':')
start_hours = int(start_split[0])
start_minutes = int(start_split[1])
start_seconds = int(start_split[2])
start_microseconds = int(start_split[3])

start = datetime.datetime(1,1,1, start_hours, start_minutes, start_seconds, start_microseconds)

finish_split = finish.split(':')

finish_hours = int(finish_split[0])
finish_minutes = int(finish_split[1])
finish_seconds = int(finish_split[2])
finish_microseconds = int(finish_split[3])

finish = datetime.datetime(1,1,1, finish_hours, finish_minutes, finish_seconds, finish_microseconds)
duration = finish - start
print("h")
duration
```
